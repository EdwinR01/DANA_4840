```python
import pandas as pd
data = pd.read_csv("data.csv")
label = pd.read_csv("labels.csv")
display(data.head())
display(label.info())
```

```python
#merge 2 dataset: "Unnamed: 0" is a shared column
merged_df = pd.merge(label,data,on='Unnamed: 0')
display(merged_df.head())
display(merged_df.info())
display(data.info())
```
